```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"D:\Data_set\dataset_olympics.csv")
```

```python
data
```

```python
sns.histplot(data = data, x ="Height",bins = 20 ,kde= True)
plt.title("Height Distribution")
plt.show()
```

```python
sns.histplot(data = data, x ='Weight',bins = 20, kde = True )
plt.title("Weight Distribution")
plt.show()
```

```python
sns.countplot(data = data, x= "Medal")
plt.title("Medal Distribution")
plt.show()
```

```python
sns.countplot(data = data, x = "Year", hue = "Medal")
plt.title("Medal Distribution Over the Year")
plt.xticks(rotation = 45)
plt.show()
```

```python
year_avg_age = data.groupby("Year")["Age"].mean()
```

```python
sport_event_count= data.groupby("Sport")["Event"].nunique().sort_values(ascending = False)
sport_event_count.plot(kind = 'bar')
plt.title("Number of Unique Events per Year")
plt.xlabel("Sports")
plt.ylabel("No. of Unique Events")
plt.xticks(rotation = 45)
plt.show()
```

```python
year_participant_count = data.groupby("Year")["ID"].nunique()
year_participant_count.plot(kind = "bar")
plt.title("Number of Participants over the year")
plt.xlabel("Year")
plt.ylabel("No. of Participants")
plt.xticks(rotation = 45)
plt.show()
```

```python
country_avg_age = data.groupby("NOC")["Age"].mean().sort_values(ascending= False)
country_avg_age.head(10).plot(kind = "bar")
plt.title("Top Ten countries with highest average level of participants")
plt.xlabel("Country")
plt.ylabel("Average Age")
plt.xticks(rotation = 45)
plt.show()
```

```python
sns.boxplot(data = data, x = "Season", y = "Age")
plt.title("Distribution of Age by Season")
plt.xlabel("Season")
plt.ylabel("Age")
plt.show()
          
```

```python
sns.violinplot(data = data, x = "Medal", y = "Height", palette = "Set2")
plt.title("Distribution of Height by medal")
plt.xlabel("Medal")
plt.ylabel("Height")
plt.show()
```

```python
most_medal_country = data["NOC"].value_counts().idxmax()
print("Most Medal winning countries : ", most_medal_country)
```

```python
tallest_athelet = data[data["Height"]== data["Height"].max()]
print("Tallest_Athelet:")
print(tallest_athelet[[   "ID","Height","Name","Sport"]])
```

```python
heaviest_athelet = data[data["Weight"]== data["Weight"].max()]
print("Heaviest_Athelet:")
print(heaviest_athelet[["Name","ID","Weight","Sport"]])
```

```python
sns.scatterplot(data = data , x = "Height", y = "Weight", hue ="Medal", palette = "Set2")
plt.title("Athelet Height vs Weight by medal status")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.legend(title = "Medal")
plt.show()
```

```python
medals_by_country_year = data.pivot_table(index = "NOC", columns= "Year",values = "Medal", aggfunc = "count")
sns.heatmap(medals_by_country_year,
            linewidth = 0.5)
plt.title("Medal Counts By aCountry and Year")
plt.xlabel("Year")
plt.ylabel("Country")
plt.xticks(rotation = 45)
plt.show()
```
